```python
import matplotlib.pyplot as plt
from math import pi as pie
import numpy as np

rad2deg  = 180/pie
# Import logs
jsdata_jimp = np.load('js_data_jointimp.npy')
tsdata_jimp = np.load('ts_data_jointimp.npy')
jsdata_timp = np.load('js_data_taskimp.npy')
tsdata_timp = np.load('ts_data_taskimp.npy')
```

```python
ts_color = 'royalblue'
js_color = 'deepskyblue'
```

```python
# Taskspace Impedance Controller
t_offset = 700      # samples [dt = 0.001 s]
time =   tsdata_timp[t_offset:,0]   # [s]
ee_pos = tsdata_timp[t_offset:,1]   # [m]
ee_vel = tsdata_timp[t_offset:,2]   # [m/s]
ee_int = tsdata_timp[t_offset:,3]   # [N]
ee_acc = tsdata_timp[t_offset:,4]   # [m/ss]

i3e_column_width = 3.5*2 # [in]
aspect_ratio = 9/16
plt.figure(figsize=[i3e_column_width, aspect_ratio*i3e_column_width])

ax1 = plt.subplot(2, 3, 1)
plt.plot(ee_pos, ee_int, color=ts_color)
plt.tick_params('y', labelsize=10)
plt.ylabel('$f_{ext}$',size='large', math_fontfamily='cm')

plt.subplot(2, 3, 2, sharey=ax1)
plt.plot(ee_vel, ee_int, color=ts_color)
plt.tick_params('y', labelleft=False)

plt.subplot(2, 3, 3, sharey=ax1)
plt.plot(ee_acc, ee_int, color=ts_color)
plt.tick_params('y', labelleft=False)

plt.subplot(2, 3, 4)
plt.plot(rad2deg*jsdata_timp[t_offset:, 1], jsdata_timp[t_offset:, 7], color=js_color)

plt.subplot(2, 3, 5)
plt.plot(rad2deg*jsdata_timp[t_offset:, 2], jsdata_timp[t_offset:, 8], color=js_color)

plt.subplot(2, 3, 6)
plt.plot(rad2deg*jsdata_timp[t_offset:, 3], jsdata_timp[t_offset:, 9], color=js_color)

plt.tight_layout()
```
