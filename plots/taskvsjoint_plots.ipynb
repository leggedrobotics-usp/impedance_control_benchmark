```python
import matplotlib.pyplot as plt
from math import pi as pie
import numpy as np

rad2deg  = 180/pie
# Import logs
jsdata_jimp = np.load('js_data_jointimp.npy')
tsdata_jimp = np.load('ts_data_jointimp.npy')
jsdata_timp = np.load('js_data_taskimp.npy')
tsdata_timp = np.load('ts_data_taskimp.npy')
```

```python
ts_color = '#6baed6'
js_color = '#2171b5'
font = {'family': 'serif', 'math_fontfamily': 'cm',
        'color': 'black', 'weight': 'semibold', 'size': 18}

i3e_column_width = 3.5*2 # [in]
aspect_ratio = 3/4
```

```python
# Task-space Impedance Controller
t_offset = 700      # samples [dt = 0.001 s]
time =   tsdata_timp[t_offset:,0]   # [s]
ee_pos = tsdata_timp[t_offset:,1]   # [m]
ee_vel = tsdata_timp[t_offset:,2]   # [m/s]
ee_int = tsdata_timp[t_offset:,3]   # [N]
ee_acc = tsdata_timp[t_offset:,4]   # [m/ss]

plt.figure()    # figsize=[i3e_column_width, aspect_ratio*i3e_column_width]

ax1 = plt.subplot(2, 3, 1, xticklabels=[], yticklabels=[])
plt.plot(ee_pos, ee_int, color=ts_color)
plt.tick_params('y', labelsize=10)
plt.ylabel('$f_{ext}$', fontdict=font)
plt.xlabel('$e_x$', fontdict=font)

plt.subplot(2, 3, 2, sharey=ax1, xticklabels=[], yticklabels=[])
plt.plot(ee_vel, ee_int, color=ts_color)
plt.xlabel('$\dot{e}_x$', fontdict=font)
#plt.tick_params('y', labelleft=False)

plt.subplot(2, 3, 3, sharey=ax1, xticklabels=[], yticklabels=[])
plt.plot(ee_acc, ee_int, color=ts_color)
plt.xlabel('$\ddot{e}_x$', fontdict=font)
#plt.tick_params('y', labelleft=False)

plt.subplot(2, 3, 4, xticklabels=[], yticklabels=[])
plt.plot(rad2deg*jsdata_timp[t_offset:, 1], jsdata_timp[t_offset:, 7], color=js_color)
plt.ylabel('$\\tau_i$', fontdict=font)
plt.xlabel('$q_1$', fontdict=font)

plt.subplot(2, 3, 5, xticklabels=[], yticklabels=[])
plt.plot(rad2deg*jsdata_timp[t_offset:, 2], jsdata_timp[t_offset:, 8], color=js_color)
plt.xlabel('$q_2$', fontdict=font)

plt.subplot(2, 3, 6, xticklabels=[], yticklabels=[])
plt.plot(rad2deg*jsdata_timp[t_offset:, 3], jsdata_timp[t_offset:, 9], color=js_color)
plt.xlabel('$q_3$', fontdict=font)

plt.subplots_adjust(left=0.10, bottom=0.10, right=0.9, top=0.90, wspace=0.05, hspace=0.26)
#plt.tight_layout()
plt.savefig('taskspace.eps')
```
